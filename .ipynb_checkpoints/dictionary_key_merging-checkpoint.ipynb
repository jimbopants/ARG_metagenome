### Gene names and other info from existing description file


```python
# imports
import pandas as pd
import numpy as np

# filepaths
gene_description = "AR_genes_clean/for_example"
accession_category_file_from_dict = "BF1_nt_dict/for_example"
combined_excel = "some/path.xlsx"
```

```python
# 1. Split AR_genes file back into a dictionary with keys that match the sorted_dict. 
names_dict = {}
with open(gene_description, 'r') as f:
    for line in f:
        name_val =  line[1:].split(" ", 1)
        names_dict[name_val[0]] = name_val[1]

        
# 2. Add gene names + metadata to dict sorted columns.
df = pd.DataFrame(columns = ['geneid', 'accession', 'category', 'name', 'encodes', 'organism', 'full description'])
with open(accession_category_file_from_dict, 'r') as f:
    for line in f:
        # Split sorted dict file into columns
        g_id = line.split()[0]
        accession = line.split()[1]
        category = line.split("\t")[2][:-1]
        
        # Parse gene description
        values = names_dict[accession]
        name = values.split(". ",1)[0]
        species = values.split("[")[1][:-2]
        
        # I attempted to parse the description so that if it has an "encodes x". I split that out as a column. 
        description = values.split(" ",1)[1]
        encodes = ''
        # Sometimes this fails because the descriptin is misses this so I put it into a try block
        try:
            encodes = description.split("ARO",1)[0]
            if len(encodes.split("encodes ")) >1:
                encodes = encodes.split("encodes ")[1][:-2]
            else:
                encodes = ''
        except:
            pass
        
        # save everything from that accession and add it as a row to the dataframe.
        srs = pd.Series(index=df.columns, data=[g_id, accession, category, name,encodes, species, description])
        df = df.append(srs, ignore_index=True)

# Save as an excel file
df.to_excel(combined_excel)
```
